```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_base_264 = pd.read_csv("csvfiles/base_264_cmd.csv")
df_base_265 = pd.read_csv("csvfiles/base_265_cmd.csv")
crf51_265 = pd.read_csv("csvfiles/crf51_265_cmd.csv")
crf51_264 = pd.read_csv("csvfiles/crf51_264_cmd.csv")
fastdecode_264 = pd.read_csv("csvfiles/fastdecode_264_cmd.csv")
fastdecode_265 = pd.read_csv("csvfiles/fastdecode_265_cmd.csv")
grain_264 = pd.read_csv("csvfiles/grain_264_cmd.csv")
grain_265 = pd.read_csv("csvfiles/grain_265_cmd.csv")
slow_264 = pd.read_csv("csvfiles/slow_264_cmd.csv")
slow_265 = pd.read_csv("csvfiles/slow_265_cmd.csv")
veryfast_264 = pd.read_csv("csvfiles/veryfast_264_cmd.csv")
veryfast_265 = pd.read_csv("csvfiles/veryfast_265_cmd.csv")
veryslow_264 = pd.read_csv("csvfiles/veryslow_264_cmd.csv")
veryslow_265 = pd.read_csv("csvfiles/veryslow_265_cmd.csv")
zeroletency_264 = pd.read_csv("csvfiles/zerolatency_264_cmd.csv")
zerolatency_265 = pd.read_csv("csvfiles/zerolatency_265_cmd.csv")
```

```python
data = df_base_264[['file_name', 'video_length', 'resolution','bit_depth','frame_rate']].copy()
```

```python
variable_cols = ['file_size', 'avg_rgb', 'bit_rate', 'psnr', 'ssim', 'encoding_time']
prefix_names = ['base_264', 'base_265', 'crf51_264', 'crf51_265', 'fastdecode_264', 'fastdecode_265', 'grain_264', 'grain_265', 'slow_264', 'slow_265', 'veryfast_264', 'veryfast_265', 'veryslow_264', 'veryslow_265', 'zeroletency_264', 'zerolatency_265']
df_list = [df_base_264,df_base_265,crf51_264,crf51_265,fastdecode_264,fastdecode_265,grain_264,grain_265,slow_264,slow_265,veryfast_264,veryfast_265,veryslow_264,veryslow_265,zeroletency_264,zerolatency_265]
```

```python
for df, prefix_name in zip(df_list, prefix_names):
    for col in variable_cols:
        df[f'{prefix_name}_{col}'] = df[col]
        df.drop(columns=[col], inplace=True)
        data = pd.concat([data, df[f'{prefix_name}_{col}']], axis=1)
```

```python

data['base_264_file_size'] = data['base_264_file_size']/1024
data['crf51_264_file_size'] = data['crf51_264_file_size']/1024
data['fastdecode_264_file_size'] = data['fastdecode_264_file_size']/1024
data['grain_264_file_size'] = data['grain_264_file_size']/1024
data['slow_264_file_size'] = data['slow_264_file_size']/1024
data['veryfast_264_file_size'] = data['veryfast_264_file_size']/1024
data['veryslow_264_file_size'] = data['veryslow_264_file_size']/1024
data['zeroletency_264_file_size'] = data['zeroletency_264_file_size']/1024


# Mean file size of each encoding presets in H265

data['base_265_file_size'] = data['base_265_file_size']/1024
data['crf51_265_file_size'] = data['crf51_265_file_size']/1024
data['fastdecode_265_file_size']   = data['fastdecode_265_file_size']/1024
data['grain_265_file_size'] = data['grain_265_file_size']/1024
data['slow_265_file_size'] = data['slow_265_file_size']/1024
data['veryfast_265_file_size'] = data['veryfast_265_file_size']/1024
data['veryslow_265_file_size'] = data['veryslow_265_file_size']/1024
data['zerolatency_265_file_size'] = data['zerolatency_265_file_size']/1024

```

```python
import os
# Get original sizes
sizes = []
for video in os.listdir('videos'):
    sizes.append(os.path.getsize(f'videos/{video}'))
sizes = [size/1024 for size in sizes]
data['original_size'] = sizes
data.head()
```

```python
# find compression ratio of each category
data['base_264_ratio'] = data['original_size']/data['base_264_file_size']
data['crf51_264_ratio'] = data['original_size']/data['crf51_264_file_size']
data['fastdecode_264_ratio'] = data['original_size']/data['fastdecode_264_file_size']
data['grain_264_ratio'] = data['original_size']/data['grain_264_file_size']
data['slow_264_ratio'] = data['original_size']/data['slow_264_file_size']
data['veryfast_264_ratio'] = data['original_size']/data['veryfast_264_file_size']
data['veryslow_264_ratio'] = data['original_size']/data['veryslow_264_file_size']
data['zeroletency_264_ratio'] = data['original_size']/data['zeroletency_264_file_size']

data['base_265_ratio'] = data['original_size']/data['base_265_file_size']
data['crf51_265_ratio'] = data['original_size']/data['crf51_265_file_size']
data['fastdecode_265_ratio'] = data['original_size']/data['fastdecode_265_file_size']
data['grain_265_ratio'] = data['original_size']/data['grain_265_file_size']
data['slow_265_ratio'] = data['original_size']/data['slow_265_file_size']
data['veryfast_265_ratio'] = data['original_size']/data['veryfast_265_file_size']
data['veryslow_265_ratio'] = data['original_size']/data['veryslow_265_file_size']
data['zerolatency_265_ratio'] = data['original_size']/data['zerolatency_265_file_size']
```

```python
data.to_csv('combined_data.csv', index=False)
```
